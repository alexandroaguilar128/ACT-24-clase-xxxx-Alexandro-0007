### Descripcion
🏥 Modelado de la Clase: Paciente
La clase Paciente actúa como un molde para crear objetos que representen a las personas que reciben atención médica.

🧬 Estructura de Datos (Atributos)
Los atributos definen el estado o la información que almacenamos de cada paciente:

1. nombre: Cadena de texto con el nombre completo.

2. edad: Valor entero que representa los años del paciente.

3. enfermedad: Diagnóstico actual del paciente.

4. tipoSangre: Grupo sanguíneo (ej. "O+", "A-").

⚙️ Comportamiento (Métodos)
Las funciones definen las acciones que el paciente puede realizar o que se pueden ejecutar sobre él:

* agendarCita(): Gestiona la asignación de una fecha para consulta.

* recetarMedicamento(): Procesa la prescripción basada en la enfermedad.

* realizarExamen(): Ejecuta pruebas diagnósticas específicas.

```python
class Paciente:
    def __init__(self, nombre, edad, enfermedad, tipo_sangre):
        self.nombre = nombre
        self.edad = edad
        self.enfermedad = enfermedad
        self.tipo_sangre = tipo_sangre

    def agendar_cita(self):
        print(f"\n--- Gestión de Cita para {self.nombre} ---")
        fecha = input("¿Para qué fecha desea la cita?: ")
        print(f"✅ Confirmado: Cita programada para el {fecha}.")

    def recetar_medicamento(self):
        print(f"\n--- Prescripción Médica ---")
        print(f"El paciente padece de: {self.enfermedad}")
        med = input("Indique el medicamento a recetar: ")
        print(f"✅ Registrado: Se ha recetado {med} a {self.nombre}.")

    def realizar_examen(self):
        print(f"\n--- Orden de Laboratorio ---")
        examen = input(f"¿Qué examen requiere el paciente (Sangre {self.tipo_sangre})?: ")
        print(f"✅ Orden enviada: Procesando examen de '{examen}'.")

# --- FLUJO DE EJECUCIÓN DIRECTO ---

print("🏥 INICIO DE REGISTRO CLÍNICO")
print("Por favor, ingrese los datos básicos del paciente:")

# El usuario llena el perfil del paciente
p = Paciente(
    input("Nombre completo: "),
    input("Edad: "),
    input("Enfermedad: "),
    input("Tipo de sangre: ")
)

# El código ejecuta las funciones una tras otra de forma interactiva
p.agendar_cita()
p.recetar_medicamento()
p.realizar_examen()

print(f"\nPROCESO FINALIZADO: El expediente de {p.nombre} ha sido actualizado con éxito. 👋")
```
