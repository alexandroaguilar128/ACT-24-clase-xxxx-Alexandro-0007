### Clase **Paciente**

   - Atributos: nombre, edad, enfermedad, tipoSangre
   - Funciones: agendarCita(), recetarMedicamento(), realizarExamen()

### Descripcion
🏥 Modelado de la Clase: Paciente
La clase Paciente actúa como un molde para crear objetos que representen a las personas que reciben atención médica.

🧬 Estructura de Datos (Atributos)
Los atributos definen el estado o la información que almacenamos de cada paciente:

1. nombre: Cadena de texto con el nombre completo.

2. edad: Valor entero que representa los años del paciente.

3. enfermedad: Diagnóstico actual del paciente.

4. tipoSangre: Grupo sanguíneo (ej. "O+", "A-").

⚙️ Comportamiento (Métodos)
Las funciones definen las acciones que el paciente puede realizar o que se pueden ejecutar sobre él:

* agendarCita(): Gestiona la asignación de una fecha para consulta.

* recetarMedicamento(): Procesa la prescripción basada en la enfermedad.

* realizarExamen(): Ejecuta pruebas diagnósticas específicas.

### Infografia
![Info](https://raw.githubusercontent.com/alexandroaguilar128/ACT-24-clase-xxxx-Alexandro-0007/refs/heads/main/infoclasexxx.png)

```python
class Paciente:
    def __init__(self, nombre, edad, enfermedad, tipoSangre):
        self.nombre = nombre
        self.edad = edad
        self.enfermedad = enfermedad
        self.tipoSangre = tipoSangre

    def agendarCita(self):
        print(f"📅 Cita agendada para {self.nombre}.")

    def recetarMedicamento(self):
        print(f"💊 Medicamento recetado para tratar {self.enfermedad}.")

    def realizarExamen(self):
        print(f"🔬 Realizando examen (Tipo de sangre: {self.tipoSangre}).")

# --- Registro del Primer Paciente ---
print("--- DATOS PACIENTE 1 ---")
p1 = Paciente(input("Nombre: "), input("Edad: "), input("Enfermedad: "), input("Tipo de Sangre: "))
p1.agendarCita()
p1.recetarMedicamento()
p1.realizarExamen()

print("\n" + "-"*20 + "\n")

# --- Registro del Segundo Paciente ---
print("--- DATOS PACIENTE 2 ---")
p2 = Paciente(input("Nombre: "), input("Edad: "), input("Enfermedad: "), input("Tipo de Sangre: "))
p2.agendarCita()
p2.recetarMedicamento()
p2.realizarExamen()
print("Programa realizado por Alexandro Aguilar NL = 1")
```
